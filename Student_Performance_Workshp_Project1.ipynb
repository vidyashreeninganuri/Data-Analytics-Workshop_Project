```python
def grade(Marks):
  if Marks >= 90:
      return  "A"
  elif Marks >= 80:
      return "B"
  elif Marks >= 70:
      return "C"
  elif Marks >= 60:
      return "D"
  else:
      return "F"
clean_data["grade"] = clean_data["Percentage"].apply(grade)
print(clean_data)
```
